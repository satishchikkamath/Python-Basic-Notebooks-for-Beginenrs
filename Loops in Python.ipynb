# For loop Using range() function
#Syntax is range(start,stop,step size)  
#The start represents the beginning of the iteration.
#The stop represents that the loop will iterate till stop-1. 
#The range(1,5) will generate numbers 1 to 4 iterations. It is optional.
#The step size is used to skip the specific numbers from the iteration. 
#It is optional to use. By default, the step size is 1. It is optional.

```python
#Example-1: Program to print numbers in sequence.
for i in range(10):  
    print(i,end = ' ')  
```

```python
#Example - 2: Program to print table of given number.
n = int(input("Enter the number "))  
for i in range(1,11):  
    c = n*i  
    print(n,"*",i,"=",c)  
```

```python
list = ['Peter','Joseph','Ricky','Devansh']  
for i in range(len(list)):  
    print("Hello",list[i])  
```

```python
#Exit the loop when x is "banana":
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

```python
#Exit the loop when x is "banana", but this time the break comes before the print:
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    break
  print(x)
```

```python
#The continue Statement
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

```python
#Example- 1: Nested for loop
# User input for number of rows  
rows = int(input("Enter the rows:"))  
# Outer loop will print number of rows  
for i in range(0,rows+1):  
# Inner loop will print number of Astrisk  
    for j in range(i):  
        print("*",end = '')  
    print()  
```

```python
#Example-2: Program to number pyramid.
rows = int(input("Enter the rows"))  
for i in range(0,rows+1):  
    for j in range(i):  
        print(i,end = '')  
    print()  
```

```python
#Using else statement with for loop
for i in range(0,5):    
    print(i)    
else:  
    print("for loop completely exhausted, since there is no break.")  
```

```python
for i in range(0,5):    
    print(i)    
    break;    
else:print("for loop is exhausted");    
print("The loop is broken due to break statement...came out of the loop")    
```

# WHILE LOOP

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
#THE ELSE STATEMENT
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

```python
# prints all letters except 'T' and 'H'   
i = 0  
str1 = 'BSCKLETECH'  
  
while i < len(str1):   
    if str1[i] == 'T' or str1[i] == 'H':   
        i += 1  
        continue  
    print('Current Letter :', str1[i])   
    i += 1  
```
